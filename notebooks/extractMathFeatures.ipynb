```python
def getAreaOfPos(pos_tags, tag):
    i = 0
    size = len(pos_tags)
    big_res = []
    res = []
    while i < size:
        if pos_tags[i][1] == tag:
            if i == 0:
                res.append(('',''))
            else:
                res.append(pos_tags[i-1])
            res.append(pos_tags[i])
            if i + 1 == size:
                res.append(('',''))
            else:
                res.append(pos_tags[i+1])
            big_res.append(res)
            res = []
        i+=1
    return big_res
```

```python
def init_compound():
    return {
        'qnt': None,
        'ent': None,
        'loc': None,
        'cnt': None,
        'adj': None,
        'vrb': None
    }
```

```python
def get_parent_idx(compound):
    return int(compound[2])

def get_word(compound):
    return compound[0]

def get_pos(compound):
    return compound[1]

def get_sentence_type(compound):
    return compound[3]
```

```python
def build_ent(compound, tree, lastQset):
    parent_idx = get_parent_idx(compound['qnt'])
    if 'NN' in get_pos(tree[parent_idx - 1]):
        compound['ent'] = tree[parent_idx - 1]
    if compound['ent'] is None:
        for word in tree:
            if get_parent_idx(word) == parent_idx and \
                word != compound['ent'] and \
                'NN' in get_pos(word):
                    compound['ent'] = word
    if compound['ent'] is None and lastQset is not None:
        compound['ent'] = lastQset['ent']
```

```python
def build_cnt(compound, tree, lastQset):
    verb_idx = tree.index(compound['vrb'])
    if 'PR' in get_pos(compound['ent']) and lastQset is not None:
        compound['cnt'] = lastQset['cnt']
    elif False:
        a = 2
    else:
        for word in tree:
            if get_sentence_type(word) == 'nsubj' \
                and get_parent_idx(word) == verb_idx - 1:
                    compound['cnt'] = word
                    break
```

```python
def build_vrb(compound, tree):
    parent_idx = get_parent_idx(compound['ent'])
    pos = get_pos(tree[parent_idx - 1])
    while parent_idx != 0 and 'VB' not in pos:
        parent_idx = get_parent_idx(tree[parent_idx - 1])
        pos = get_pos(tree[parent_idx - 1])
    compound['vrb'] = tree[parent_idx - 1]
```

```python
def build_loc(compound, tree):
    idxs = [i for i in range(len(tree)) \
            if get_sentence_type(tree[i]) == 'case' and get_word(tree[i]).lower() != 'of']
    if len(idxs) > 0:
        compound['loc'] = []
        for idx in idxs:
            loc_words = [tree[idx]]
            parent_idx = get_parent_idx(tree[idx])
            while parent_idx != 0 and 'VB' not in get_pos(tree[parent_idx - 1]):
                loc_words.append(tree[parent_idx - 1])
                parent_idx = get_parent_idx(tree[parent_idx - 1])
            if tree[parent_idx - 1] == compound['vrb']:
                compound['loc'] += loc_words
```

```python
def build_adj(compound, tree):
    idx = tree.index(compound['ent'])
    adj = []
    for word in tree:
        if get_parent_idx(word) == idx + 1 and \
            get_sentence_type(word) == 'amod':
                adj.append(word)
    compound['adj'] = adj
```

```python
def build_compound(tree):
    new_compond = None
    compounds = []
    last_Qset = None
    for word in tree:
        if get_pos(word) == 'CD' or get_sentence_type(word) == 'compound' \
            or get_pos(word) == 'DT':
            new_compound = init_compound()
            new_compound['qnt'] = word
            build_ent(new_compound, tree, last_Qset)
            build_vrb(new_compound, tree)
            build_loc(new_compound, tree)
            build_cnt(new_compound, tree, last_Qset)
            build_adj(new_compound, tree)
            last_Qset = new_compound
            compounds.append(new_compound)
    return compounds
```

```python
def build_qsets(text):
    dep_trees = get_parse(text)
    compounds = []
    sent_cnt = 1
    for tree in dep_trees:
        new_compound = build_compound(tree)
        for x in new_compound:
            x['sent_cnt'] = sent_cnt
        if new_compound is not None:
            compounds.append(new_compound)
        sent_cnt+=1
    return compounds
```
